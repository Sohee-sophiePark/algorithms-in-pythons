# Tree Traversal (Depth First traversal - InOrder, PreOrder, and PostOrder / Breadth-First Traversal)
1. Inorder 
    Left-subtree -> Root -> Right-subtree
2. Preorder 
    Root -> Left-subtree -> Right-subtree
3. Postorder 
    Left-subtree -> Right-subtree -> Root

* Inorder Algorithm 
1. Traverse the left-subtree
    - call Inorder(left-subtree)
2. Visit the root
3. Traverse the right-subtree
    - call Inorder(right-subtree)


Example: BST, non-increasing order, non-decreasing order

```python
class Node:
    def __init__(self, data, prev = None, next = None):
        self.data = data
        self.prev = prev
        self.next = next

class DoubleLinkedList:
    def __init__(self):
        self.head = None


    def print_forward(self):
        if self.head is None:
            print(f'No data in Double Linked List')
            return
        
        itr = self.head
        output = ''
        while itr:
            # print(f'itr.prev {itr.prev}')
            # print(f'itr.data {itr.data}')
            # print(f'itr.next {itr.next}')
            output += str(itr.data)
            if itr.next:
                output += ' --> '

            itr = itr.next
       

        print(output)
        return output

    def print_backward(self):
        if self.head is None:
            print(f'No data in Double Linked List')
            return

        ## method 1 - better perf
        output = ''
        itr = self.head
        while itr:
            # print(f'itr.data {itr.data}')
            if itr.prev is None:
                output += str(itr.data)
            else:
                output =  str(itr.data) + ' --> ' + output           
            itr = itr.next
            
        ## method 2
        # output = self.print_forward()
        # output = ' --> '.join(reversed(output.split(' --> ')))
        
        print(output)
        

    def insert_at_beginning(self, data):
        if self.head is None:
            # print(self.head)
            self.head = Node(data, None, None)
            return
        
        self.head.prev = Node(data, None, self.head)
        node = self.head
        self.head = Node(data, None, node)
        # print(f'self.head.prev.data {(self.head.next.prev).data}')
        # print(f'self.head.data {self.head.data}')

    
    def insert_at_end(self, data):
        if self.head is None:
            self.head = Node(data, None, None)
            return

        itr = self.head
        while itr.next:
            itr = itr.next

        itr.next = Node(data, itr, None)
        # print(f'itr.prev {itr.prev.data}')
        # print(f'itr.data {itr.data}')
        # print(f'itr.next {itr.next.data}')
        
    def insert_at(self, index, data):
        if index < 0 or index > self.get_length():
            raise Exception("Invalid Bound Index")
        
        if index == 0:
            self.insert_at_beginning(data)
            return
        if index == self.get_length():
            self.insert_at_end(data)
            return

        itr = self.head
        for i in range(self.get_length()):
            if i == index-1:
                node = Node(data, itr, itr.next)
                itr.next = node
            itr = itr.next
        


    def insert_values(self, values):
        for value in values:
            self.insert_at_end(value)



    def remove_at(self, index):
        if index < 0  or index >= self.get_length():
            raise Exception("Invalid Bound Index")
        
        if index == 0:
            self.head = self.head.next
            self.head.prev = None

        itr = self.head
        for i in range(self.get_length()):
            if i == index-1:
                itr.next = itr.next.next
                return
            itr = itr.next
        

    def find_index(self, data):
        itr = self.head
        for i in range(self.get_length()):
            if itr.data == data:
                return i
            itr = itr.next
        
        print(f'{data} is not in list')
        

    def insert_by_value(self, after_value, insert_data):
        itr = self.head
        for i in range(self.get_length()):
            if itr.data == after_value:
                itr.next = Node(insert_data, itr, itr.next)
                # print(itr.data)
                # print(itr.next.data)
                # print(itr.next.next.data)
                return
            itr = itr.next
        print(f'{after_value} is not in list')


    def remove_by_value(self, data):
        itr = self.head
        for i in range(self.get_length()):
            if itr.data == data:
                itr.prev.next = itr.next
                return
            itr = itr.next
        print(f'{data} is not in list')


    # def insert_after_data(self, after_data, inserted_data):

    def print(self):
        self.print_forward()

    def get_length(self):
        counter = 0
        itr = self.head

        while itr:
            counter += 1
            itr = itr.next
        
        return counter

    
```

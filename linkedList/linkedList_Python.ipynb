```python
class Node:
    def __init__(self, data=None, next = None):
        self.data = data
        self.next = next

class LinkedList:
    def __init__(self):
        self.head = None

    def insert_at_begining(self, data):
        if self.head is None:
            self.head = Node(data, None)
            return

        node = self.head
        self.head = Node(data, node)
        

    def insert_at_end(self, data):
        if self.head is None:
            self.head = Node(data, None)
            return
            
        itr = self.head
        while itr.next:
            itr = itr.next
        
        itr.next = Node(data, None)

    def insert_values(self, data_list):
        for data in data_list:
            self.insert_at_end(data)
            

    def get_length(self):
        counter = 0
        itr = self.head
        while itr:
            counter += 1
            itr = itr.next
        
        print(f'lenght: {counter}')
        return counter

    
    def remove_at(self, index):
        if index < 0 or self.get_length() <= index:
            raise Exception("Out of bound")
        if index == 0:
            self.head = self.head.next
            return

        itr = self.head
        for i in range(self.get_length()):
            if i == index-1:
                itr.next = itr.next.next
                return
            else:
                itr = itr.next

    def insert_at(self, index, data):
        if index < 0 or index > self.get_length():
            raise Exception("Out of bound")
        if index == 0:
            self.insert_at_begining(data)
            return
        if index == self.get_length():
            self.insert_at_end(data)
            return
            
            # node = self.head
            # self.head = Node(data, node)
            # return


        itr = self.head
        for i in range(self.get_length()+1):
            print(index)
            print(i)
            if i == index-1:
                print(index)
                print(i)
                node = Node(data, itr.next)
                itr.next = node
                return
            itr = itr.next

    def insert_after_value(self, data_after, data_to_insert):
        if data_after is None or data_to_insert is None:
            print("Invalid Input")
            return
        itr = self.head
        while itr.next:
            if itr.data == data_after:
                node = Node(data_to_insert, itr.next)
                itr.next = node
                return
            itr = itr.next
        print(f'{data_after} is not in the LinkedList')


    def remove_by_value(self, data):
        if data is None:
            print("Invalid Input")
            return
        itr = self.head
        while itr.next:
            print(itr.next.data)
            if itr.next.data == data:
                itr.next = itr.next.next
                return
            itr = itr.next
        print(f"{data} is not in LinkedList")
            

    def print(self):
        if self.head is None:
            print("Empty")
        itr = self.head
        output = ''


        while itr:
            # print(itr.data, itr.next)
            output += str(itr.data)
            if itr.next:
                output += ' -> '
            itr = itr.next

        print(output)
```
